# Modeling and Validation Presentation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score, cross_val_predict, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import RidgeCV, LassoCV
```

### Run same model from Kaggle comp, however only look at the train set

```python
# import train data
train = pd.read_csv('./datasets/train_mod03.csv')
raw_train = pd.read_csv('./datasets/train.csv')

print(train.shape)
```

```python
# define features and target var
features = train.drop(columns=['SalePrice', 'PID', 'Id']).columns
X = train[features]
y = np.log(train['SalePrice'])
```

```python
# Standardize X in order to run regularized regression
ss = StandardScaler()
ss.fit(X)
Z = ss.transform(X)
```

```python
# Run ridge regression with optimized alpha hyperparameter
ridge_alphas = np.linspace(600, 800, 100)
# set model to optimize root mean squared error
ridge_cv = RidgeCV(alphas = ridge_alphas, scoring = 'neg_root_mean_squared_error', cv = 5)
model = ridge_cv.fit(Z, y)
# print alpha parameter to ensure it is not on the boundry of ridge_alphas
print(ridge_cv.alpha_)
```

```python
# model evaluation
print(cross_val_score(model, Z, y, cv=5).mean())
print(cross_val_score(model, Z, y, cv=5).std())
```

### Look at differences between predicted sales price and actual sales price

```python
#y_hat = np.exp(model.predict(Z_test))
y_hat = np.exp(model.predict(Z))

y = np.exp(y)
residuals = y - y_hat
```

```python
d = {'y': y, 'y_hat': y_hat, 'resids': residuals}
resids_df = pd.DataFrame(data=d)
```

```python
# find predicitons that more than 10% greater than actual values
resids_df['under_valued'] = resids_df['y'] * 1.1 < resids_df['y_hat']
raw_train['under_valued'] = resids_df['under_valued']
```

```python
# plot the residuals in order to visualize undervalued homes
plt.figure(figsize=(12,8))
sns.scatterplot(y_hat, residuals, hue=resids_df['under_valued'], palette=['#FFFFFF', '#ffd969'], edgecolor='black')
plt.axhline(0, color='#ffd969');
plt.ylabel('Valuation Inaccuracy ($)', fontsize=12)
plt.xlabel('Asking Price ($)', fontsize=12)
plt.title('Under Valued Homes', fontsize=20);
```

### Convert the undervalued train subset to a navigable widget

```python
#https://medium.com/@marekermk/guide-to-interactive-pandas-dataframe-representation-485acae02946

def init_datatable_mode():
    """Initialize DataTable mode for pandas DataFrame represenation."""
    import pandas as pd
    from IPython.core.display import display, Javascript

    display(Javascript("""
        require.config({
            paths: {
                DT: '//cdn.datatables.net/1.10.19/js/jquery.dataTables.min',
            }
        });

        $('head').append('<link rel="stylesheet" type="text/css" href="//cdn.datatables.net/1.10.19/css/jquery.dataTables.min.css">');
    """))

    def _repr_datatable_(self):
        """Return DataTable representation of pandas DataFrame."""
        # classes for dataframe table (optional)
        classes = ['table', 'table-striped', 'table-bordered']

        # create table DOM
        script = (
            f'$(element).html(`{self.to_html(index=False, classes=classes)}`);\n'
        )

        # execute jQuery to turn table into DataTable
        script += """
            require(["DT"], function(DT) {
                $(document).ready( () => {
                    // Turn existing table into datatable
                    $(element).find("table.dataframe").DataTable();
                })
            });
        """

        return script

    pd.DataFrame._repr_javascript_ = _repr_datatable_
```

```python
init_datatable_mode()
```
